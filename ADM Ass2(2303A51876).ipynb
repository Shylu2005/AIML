```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/diabetes.csv')
plt.hist(df['Age'], bins=40, color='orange', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#2)Create Visualizations Using Matplotlib and Seaborn
#(a) Histogram for Each Feature
import matplotlib.pyplot as plt
df.hist(figsize=(12, 8), bins=20, edgecolor="black")
plt.suptitle("Feature Distributions", fontsize=14)
plt.show()
```

```python
#(b) Scatter Plot (Glucose vs. BMI as an Example)
import seaborn as snsI
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Glucose"], y=df["BMI"], hue=df["Outcome"], palette="coolwarm")
plt.title("Scatter Plot: Glucose vs. BMI")
plt.xlabel("Glucose Level")
plt.ylabel("BMI")
plt.show()
```

```python
#(c) Correlation Heatmap
corr_matrix = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
